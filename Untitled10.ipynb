```python
# Load the patients data
patients_df = pd.read_csv('/content/drive/My Drive/patients.csv')
display(patients_df.describe())
display(patients_df.isnull().sum())
# Convert 'DOB' and 'DOD' to datetime objects
patients_df['DOB'] = pd.to_datetime(patients_df['DOB'])
patients_df['DOD'] = pd.to_datetime(patients_df['DOD'])

# Define a reference date for calculating age for living patients
reference_date = pd.to_datetime('2199-12-31') # Using a date after the latest possible DOD

# Calculate age in years using year difference
patients_df['AGE'] = (patients_df['DOD'].fillna(reference_date).dt.year - patients_df['DOB'].dt.year)

# Adjust age if the birthday hasn't occurred yet in the death/reference year
patients_df.loc[(patients_df['DOD'].fillna(reference_date).dt.month < patients_df['DOB'].dt.month) |
                ((patients_df['DOD'].fillna(reference_date).dt.month == patients_df['DOB'].dt.month) &
                 (patients_df['DOD'].fillna(reference_date).dt.day < patients_df['DOB'].dt.day)), 'AGE'] -= 1


# Filter by age
adult_patients_df = patients_df[(patients_df['AGE'] >= 16) & (patients_df['AGE'] <= 89)]

# Display filtered data
display(adult_patients_df.head())
print("Shape of the filtered DataFrame:", adult_patients_df.shape)
```

```python
# age distribution
adult_patients_df[['SUBJECT_ID','AGE']].groupby('SUBJECT_ID').max().describe()
```

```python
# gender distribution
gender_dist = adult_patients_df[['SUBJECT_ID','GENDER']].groupby('SUBJECT_ID').max().GENDER.value_counts()
print('Gender ratio (male:all): ', gender_dist[0]/gender_dist.sum())
print('Gender distribution:\n',gender_dist)
```

```python
# num of ICU stays
icu_stays_df[['SUBJECT_ID','LOS']].groupby('SUBJECT_ID').max().describe()
```

```python
# Compute the median LOS
AVG_LOS = icu_stays_df.LOS.median()
print(AVG_LOS)
```

```python
filtered_icu_stays_df['OUTTIME'] = pd.to_datetime(filtered_icu_stays_df['OUTTIME'])
filtered_icu_stays_df['INTIME'] = pd.to_datetime(filtered_icu_stays_df['INTIME'])
filtered_icu_stays_df["LOS"] = (filtered_icu_stays_df["OUTTIME"]-filtered_icu_stays_df["INTIME"])/pd.Timedelta('1 hour')
```

```python
# Merge the two dataframes on 'SUBJECT_ID'
merged_df = pd.merge(filtered_icu_stays_df, adult_patients_df, on='SUBJECT_ID')

# Calculate the correlation between LOS and AGE
correlation = merged_df['LOS'].corr(merged_df['AGE'])

print(f"The correlation between Length of Stay and Age (16-89) is: {correlation}")

# Plot the distribution
merged_df.plot("AGE","LOS",kind="scatter")
plt.xlabel('Age (years)')
plt.ylabel('Length of Stay (days)')
plt.title('Scatter Plot of Length of Stay vs. Age')
plt.show()
```

```python
patients_info_df.DIAGNOSIS = patients_info_df.DIAGNOSIS.str.replace(";"," ").str.strip()
all_diagnosis_str = patients_info_df.DIAGNOSIS.str.cat(sep=" ")

subs = {"\\":" ", "-":"", "/":"", "?":""}
for s in subs.keys():
    all_diagnosis_str = all_diagnosis_str.replace(s, subs[s])

all_diagnosis = all_diagnosis_str.split(" ")
```

```python
import collections
c = collections.Counter(all_diagnosis)
# Filter out empty strings before getting the most common terms
filtered_c = collections.Counter({k: v for k, v in c.items() if k != ''})
filtered_c.most_common(10)
```

```python
top_10_keyword = sorted(c, key=c.get, reverse=True)[:10]
print(f"Top 10 most common diagnostic terms are: {top_10_keyword}")
```

```python
for keyword in top_10_keyword:
    patients_info_df[keyword]=patients_info_df['DIAGNOSIS'].str.contains(keyword)
```
